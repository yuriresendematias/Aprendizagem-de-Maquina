#### 5. Avalie o RMSE e o R2 do conjunto de treino convertido utilizando 10-fold cross validation.


```python
predict_t = model.predict(train)
predict_t
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

kf = KFold(n_splits=10, shuffle=True, random_state=2)

for train_index, test_index in kf.split(sp):
    sp_train, sp_test = sp[train_index], sp[test_index]
    predict_train, predict_test = predict_t[train_index], predict_t[test_index]

rmse2 = np.sqrt(mean_squared_error(sp_train, predict_train))
print(f'RMSE: {rmse2:.0f}')

r3 = r2_score(sp_train, predict_train)
print(f'R2: {r3:.4f}')
```

#### 6. Faça o mesmo da questão anterior utilizando regressão passo-a-passo.

```python
from dmba import stepwise_selection
from dmba import AIC_score

def train_model(variables):
    if len(variables) == 0:
        return None
    model = LinearRegression()
    model.fit(train[variables], sp)
    return model

def score_model(model, variables):
    if len(variables) == 0:
        return AIC_score(sp, [sp.mean()] * len(sp), model, df=1)
    return AIC_score(sp, model.predict(train[variables]), model)

best_model, best_variables = stepwise_selection(train.columns, train_model, score_model, verbose=True)

best_model
```

```python
stepward_train = best_model.predict(train[best_variables])
stepward_train
```

```python
for train_index, test_index in kf.split(sp):
    sp_train, sp_test = sp[train_index], sp[test_index]
    predict_train, predict_test = stepward_train[train_index], stepward_train[test_index]

rmse2 = np.sqrt(mean_squared_error(sp_train, predict_train))
print(f'RMSE: {rmse2:.0f}')

r3 = r2_score(sp_train, predict_train)
print(f'R2: {r3:.4f}')
```

#### 7. Faça o mesmo da questão anterior utilizando regressão com seleção progressiva.

```python
from dmba import forward_selection

forward_model, forward_variables = forward_selection(train.columns, train_model, score_model, verbose=True)
```

```python
forward_train = forward_model.predict(train[forward_variables])
forward_train
```

```python
for train_index, test_index in kf.split(sp):
    sp_train, sp_test = sp[train_index], sp[test_index]
    predict_train, predict_test = forward_train[train_index], forward_train[test_index]

rmse2 = np.sqrt(mean_squared_error(sp_train, predict_train))
print(f'RMSE: {rmse2:.0f}')

r3 = r2_score(sp_train, predict_train)
print(f'R2: {r3:.4f}')
```

#### 8. Faça o mesmo da questão anterior utilizando regressão com eliminação regressiva.


```python
from dmba import backward_elimination

backward_model, backward_variables = backward_elimination(train.columns, train_model, score_model, verbose=True)
```

```python
backward_train = backward_model.predict(train[backward_variables])
backward_train
```

```python
for train_index, test_index in kf.split(sp):
    sp_train, sp_test = sp[train_index], sp[test_index]
    predict_train, predict_test = backward_train[train_index], backward_train[test_index]

rmse2 = np.sqrt(mean_squared_error(sp_train, predict_train))
print(f'RMSE: {rmse2:.0f}')

r3 = r2_score(sp_train, predict_train)
print(f'R2: {r3:.4f}')
```
