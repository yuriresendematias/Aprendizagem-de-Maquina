#### 1. Identique as variáveis categóricas na base de dados e descreva como transformar apropriadamente cada uma delas para a regressão linear múltipla.

```python
import pandas as pd

data = pd.read_csv("train.csv")
cat = []

print("Variaveis categoricas") 

for colum in data.columns:
    if data[colum].dtype == "object":
        cat.append(colum)

pd.DataFrame(cat, columns=["Categoricas"])
```
